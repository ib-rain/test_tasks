Выберете, по вашему мнению, самый эффективный
- канал (`media_channel`);
- географию (`country_code`);
- платформу (`platform`).

```python
categories = ('media_channel', 'country_code', 'platform_id')
# category = categories[0]

for category in categories:
    print(in_app_events[category].value_counts())
    print(costs[category].value_counts())
```

```python
def calc_metrics(category='campaign'):
    # Переиспользуем расчёты для произвольных категорий.
    
    number_of_unique_installations = (in_app_events
        [in_app_events['event_name'] == 'install']
        [category]
        .value_counts()
        .sort_index()
        .rename('number_of_unique_installations')
    )
    
    number_of_unique_payers = (in_app_events
        [in_app_events['event_name'] == 'purchase']
        .groupby(category)
        ['customer_user_id']
        .nunique()
        .rename('number_of_unique_payers')
    )
    
    total_campaign_costs = (costs
        .groupby(category)
        ['cost']
        .sum()
        .rename('total_campaign_costs')
    )
    
    total_campaign_revenue = (in_app_events
        [in_app_events['event_name'] == 'purchase']
        .groupby(category)
        ['net_event_revenue_usd']
         .sum()
        .rename('total_campaign_revenue')
    )
    
    total_campaign_payment_count = (in_app_events
        [in_app_events['event_name'] == 'purchase']
        .groupby(category)
        ['customer_user_id']
        .count()
        .rename('total_campaign_payment_count')
    )
    
    combined_campaign = pd.concat(
        objs=(
            number_of_unique_installations,
            number_of_unique_payers,
            total_campaign_costs,
            total_campaign_revenue,
            total_campaign_payment_count
        ),
        axis=1
    )

    combined_campaign['cv'] = combined_campaign['number_of_unique_payers'] / combined_campaign['number_of_unique_installations']
    combined_campaign['roi'] = (combined_campaign['total_campaign_revenue'] - combined_campaign['total_campaign_costs']) / combined_campaign['total_campaign_costs']
    combined_campaign['cac_inst'] = -combined_campaign['total_campaign_costs'] / combined_campaign['number_of_unique_installations']
    combined_campaign['cac_pay'] = -combined_campaign['total_campaign_costs'] / combined_campaign['number_of_unique_payers']
    combined_campaign['arpu'] = combined_campaign['total_campaign_revenue'] / combined_campaign['number_of_unique_payers']
    combined_campaign['apcpu'] = combined_campaign['total_campaign_payment_count'] / combined_campaign['number_of_unique_payers']
    combined_campaign['aov'] = combined_campaign['total_campaign_revenue'] / combined_campaign['total_campaign_payment_count']
    
    return combined_campaign
```

```python
res_dict = {}
for category in categories:
    res_dict[category] = calc_metrics(category)
```
